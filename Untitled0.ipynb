```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
dataset = sns.load_dataset('penguins')
dataset.info()
```

```python
clean_data = dataset.dropna()
```

```python
sns.countplot(data=clean_data, x='island', hue='island')
plt.title('count of island')
plt.show()
```

demonstrate the scatter plot which compares  to column in the dataset

```python
sns.histplot(data=clean_data, x='gender', y='number', hue="gender")
plt.title('histplot')
```

```python
sns.barplotplot(data=clean_data, x='island', hue="island")
plt.title('barplot of   island')
plt.show()
```
